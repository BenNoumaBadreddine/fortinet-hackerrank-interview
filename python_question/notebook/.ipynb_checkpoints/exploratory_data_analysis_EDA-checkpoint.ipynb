## Univariate Analysis

The current project is a classification problem, that is why we start with the analyzing the target variable.
This section is meant to analyze each variable seperatly.

* **For categorical features** we can use frequency table or bar plots which will calculate the number of each category in a particular variable.

* **For numerical features**, probability density plots can be used to look at the distribution of the variable.

### The target variable 

In our case, the target column is : **Loan_Status**

```python
target = "Loan_Status"
```

#### Frequency

```python
# Normalize can be set to True to print proportions instead of number 
frequencies = data[target].value_counts(normalize=True)*100
frequencies
```

#### bar plot 

```python
color=['blue','orange','black', 'red', 'green',  'cyan']
data[target].value_counts(normalize=True).plot.bar(title = target, color=color[:len(frequencies)])
```

**Interpretation**: The loan approval frequency is 68.72% and the rejected loan frewquency is 31.27%
