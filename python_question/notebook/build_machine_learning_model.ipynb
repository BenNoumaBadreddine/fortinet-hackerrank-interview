# Build a machine learning model


# Load the data

```python
import sys
sys.path.append('/Users/badr/github_projects')
from sklearn.model_selection import train_test_split
from fortinet_hacker_rank_interview.python_question.src.data_preprocessing import data_preprocessing_workflow
from fortinet_hacker_rank_interview.python_question.src.building_models import build_classification_model, create_roc_curve


data = data_preprocessing_workflow()
features =['Education', 'Loan_Amount_Term', 'Credit_History', 'Loan_Status', 'Gender_Female',
  'Married_No', 'Property_Area_Rural', 'Property_Area_Semiurban',
  'Dependents_0', 'Dependents_1', 'Dependents_2', 'Self_Employed_No',
  'LoanAmount_log', 'total_income_log', 'EMI', 'Balance_Income']
data = data[features]
target = "Loan_Status"
X = data.drop(target, axis=1, inplace=False)
y = data[[target]]
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.2, random_state=1)

```

```python
x_train = x_train.values
y_train = y_train[target].values

x_test = x_test.values
y_test = y_test[target].values

x_valid = x_valid.values
y_valid = y_valid[target].values
```

# Logistic Regression

```python
results = build_classification_model(x_train, y_train, x_test, y_test, x_valid, y_valid,
                                     model_name="Logistic Regression", grid_search=True)
create_roc_curve(y_valid, results.get("valid_score_results").get("predictions"))
```

# Decision Tree

```python
results = build_classification_model(x_train, y_train, x_test, y_test, x_valid, y_valid,
                                     model_name="Decision Tree", grid_search=True)
create_roc_curve(y_valid, results.get("valid_score_results").get("predictions"))
```

# Random Forest

```python
results = build_classification_model(x_train, y_train, x_test, y_test, x_valid, y_valid,
                                     model_name="Random Forest", grid_search=False)
create_roc_curve(y_valid, results.get("valid_score_results").get("predictions"))
```

# Random Forest with Grid-search


```python
results = build_classification_model(x_train, y_train, x_test, y_test, x_valid, y_valid,
                                     model_name="Random Forest", grid_search=True)
create_roc_curve(y_valid, results.get("valid_score_results").get("predictions"))
```
